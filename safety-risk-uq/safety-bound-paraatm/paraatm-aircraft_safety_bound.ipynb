# Import IFF Data for flight ABC123 using para-atm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from paraatm.io.iff import read_iff_file
data = read_iff_file('IFF_SFO_ASDEX_ABC123.csv')
```

# Compute Instantaneous Aircraft Acceleration

```python
data['delta_tas'] = data['tas'].diff()
data['dt'] = (data['time']-data['time'].shift()).fillna(pd.to_timedelta(0.01,'s')).dt.total_seconds()
data['acc']= data['delta_tas']/data['dt'] #knots/second
```

```python
plt.plot(data.acc.values)
plt.xlabel('Time (seconds)')
plt.ylabel('Acceleration (knots/s)')
plt.grid()
```

```python
plt.plot(data.tas.values)
plt.xlabel('Time (seconds)')
plt.ylabel('Total Air Speed (knots)')
plt.grid()
```

```python
plt.plot(data.coord1Accur.values)
plt.xlabel('Time (seconds)')
plt.ylabel('Positional Accuracy')
plt.grid()
```

```python
from SafetyBound import BoundAircraft
#from paraatm.safety.SafetyBound import BoundAircraft

lhs = []
lps = []
rs = []
for i in range(len(data)-1):

    dt = pd.Timedelta(data.time[i+1]-data.time[i]).total_seconds()
    tas = data.tas[i+1]
    acc = data.acc[i+1]
    hdg = data.heading[i+1]
    posAccur = data.coord1Accur[i+1]
    vwind0 = 0.1
    alphaw0 = 0.1

    b = BoundAircraft(tas,acc,dt,posAccur,hdg,vwind0,alphaw0)
    b.size()
    lh = b.lh
    lp = b.lp
    r = b.r

    accs.append(acc)
    lhs.append(lh)
    lps.append(lp)
    rs.append(r)
```

```python
plt.plot(rs)
plt.xlabel('Time (seconds)')
plt.ylabel('Safety Bound Radius')
plt.grid()
```
